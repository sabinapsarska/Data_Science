# **K Means Clustering Algorithm**

```python
# https://scikit-learn.org/stable/auto_examples/datasets/plot_iris_dataset

import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d  # noqa: F401
from sklearn import datasets
from sklearn.decomposition import PCA

iris = datasets.load_iris()
X = iris.data[:, 2:] # only petal length and petal width
y = iris.target


# Orginal data with target
plt.figure(2, figsize=(8, 6))
plt.clf()

plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1, edgecolor="k")
plt.xlabel("petal length")
plt.ylabel("petal width")

plt.show()
```

```python
from sklearn.cluster import KMeans

sse = []
num = range(1,10)

for i in num:
  km = KMeans(n_clusters = i)
  y_predicted = km.fit_predict(X)
  sse.append(km.inertia_)

plt.plot(num, sse)
```

```python
km = KMeans(n_clusters = 3)

y_predicted = km.fit_predict(X)
y_predicted
```

```python
plt.figure(2, figsize=(8, 6))
plt.clf()

plt.scatter(X[:, 0], X[:, 1], c=y_predicted, cmap=plt.cm.Set1, edgecolor="k")
plt.xlabel("petal length")
plt.ylabel("petal width")

plt.show()
```

```python
sum = 0 
for i in range(len(y)):
  if y_predicted[i] == 1: y_predicted[i] = 0 
  elif y_predicted[i] == 2: y_predicted[i] = 1
  elif y_predicted[i] == 0: y_predicted[i] = 2

  if y[i] == y_predicted[i]: sum=sum+1

sol = sum/len(y)
sol
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(X)
X = scaler.transform(X)

km = KMeans(n_clusters = 3)
y_predicted = km.fit_predict(X)

plt.figure(2, figsize=(8, 6))
plt.clf()

plt.scatter(X[:, 0], X[:, 1], c=y_predicted, cmap=plt.cm.Set1, edgecolor="k")
plt.xlabel("petal length")
plt.ylabel("petal width")

plt.show()
```
